```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import GradientBoostingRegressor
```

```python
feature_X_df = df[['year', 'weekofyear', 'reanalysis_dew_point_temp_k',
       'reanalysis_min_air_temp_k', 'reanalysis_specific_humidity_g_per_kg',
       'station_max_temp_c', 'station_min_temp_c', 'station_precip_mm']]
feature_y_df = df[['total_cases']]
```

```python
feature_X_df  = feature_X_df .interpolate(method="linear")
```

```python
train_X, test_X, train_y, test_y = train_test_split(feature_X_df, feature_y_df,
                      test_size = 0.3, random_state = 50)
```

```python
params = {
    "n_estimators": 200,
    "max_depth": 10,
    "min_samples_split":5,
    "learning_rate": 0.02,
    "criterion": "mse"
}
```

```python
reg = GradientBoostingRegressor(**params)
reg.fit(train_X, train_y)
```

```python
y_pred = reg.predict(train_X)
```

```python
 mean_absolute_error(train_y, y_pred)
```

```python
y_pred = reg.predict(test_X)
mean_absolute_error(test_y, y_pred)
```
